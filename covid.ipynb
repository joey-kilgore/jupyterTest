## Load Data
We can access the Our World In Data Covid-19 dataset online through their github repo and the publically available csv file. For more information checkout the github repo https://github.com/owid/covid-19-data/tree/master/public/data

```r
covid <- read.csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")
USA <- covid[covid$location=="United States",]
```

The data set gives a large selection of deaths, test, cases, etc. and allows us to look at other factors within the data set.

```r
head(USA)
```

## Visualizing the Data
We can do a holistic graph to get an initial overview of a large amount of data.
The following graph includes the daily new:  
 - cases (y-axis)  
 - tests (size)  
 - deaths (color)  
    
With this we get the following graph

```r
cols <- c("blue","red")
basicUSA <- ggplot(data=USA)+
                geom_point(aes(x=as.Date(date),y=new_cases,color=new_deaths,size=new_tests))+
                scale_colour_gradientn(colours = cols)+
                theme(axis.text.x = element_text(angle = 45, hjust = 1))

basicUSA
```
